```python
# Separar Dados de Entrada (Atributos) e de saída (target)
y = df.Response
X = df.drop('Response', axis=1)
```

```python
#verificando os valores dos atributos

X.head()
```

```python
#verificando os falores de saida 

y.head()
```

```python
# Tratar colunas categóricas
y.value_counts()
```

```python
X.info()
```

```python
# Tratar colunas categoricas da entrada (e categorias ausentes)
X = pd.get_dummies(X)
```

```python
# Separação Treino/Teste
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    random_state=42,
                                                    test_size=0.2,
                                                    stratify=y)
```

```python
# Tratar dados numericos ausentes
num_cols = ['ID', 'Year_Birth', 'Kidhome', 'Teenhome', 'Recency',
            'MntWines','MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts',
           'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases',
           'NumStorePurchases', 'NumWebVisitsMonth', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5',
           'AcceptedCmp1', 'AcceptedCmp2', 'Complain', 'Z_CostContact', 'Z_Revenue']
train_median = X_train[num_cols].median()
X_train[num_cols]=X_train[num_cols].fillna(train_median)
X_test[num_cols]=X_test[num_cols].fillna(train_median)
```

```python
#verificando como ficou os atributos de treino

X_train.head()
```

```python
#utilizando a biblioteca dummy classifier


from sklearn.dummy import DummyClassifier
dummy_clf = DummyClassifier(strategy="most_frequent")
dummy_clf.fit(X_train, y_train)

y_pred = dummy_clf.predict(X_test)
```

```python
# Acuracia
np.mean(y_pred==y_test)
```

```python
#utilizando a biblioteca de metrics para ver o score da accuracy

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
#visualizar se existem valores missing

df.isnull()
```

```python
#correlação

df.corr()
```

```python
#verificando como ficou o modelo com a matrix da confusion

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='g')
plt.xlabel('Predição( 1 ou 0)')
plt.ylabel('Real ( 1 ou 0)')
```

```python
#modelo classification report

from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
#utilizando a biblioteca metrica para iportar roc auc score

from sklearn.metrics import roc_auc_score
```

```python
#utilizando roc auc

y_proba = dummy_clf.predict_proba(X_test)
roc_auc_score(y_test, y_proba[:, 1])
```

```python
#utilizando a biblioteca sklearn para importar a regression logistic

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
# Acuracia
lr.score(X_test, y_test)
```

```python
#utilizando o roc auc

y_proba = lr.predict_proba(X_test)
roc_auc_score(y_test, y_proba[:, 1])
```

```python
#biblioteca sklearn para usar o modelo de randomforestclassifier e decisiontreeclassifier

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#lista dos classificadores

lista_classificacores = [DecisionTreeClassifier(),
                         RandomForestClassifier(),]
```

```python
#treinando os modelos, a acuracia, teste

for clf in lista_classificacores:
  print(clf.__class__.__name__)
  clf.fit(X_train, y_train)

  print('Acuracia')
  train_acc=clf.score(X_train, y_train)
  test_acc=clf.score(X_test, y_test)
  print('Treino:', train_acc)
  print('Teste:', test_acc)

  print('ROC AUC')
  y_proba_train = clf.predict_proba(X_train)
  roc_train=roc_auc_score(y_train, y_proba_train[:, 1])
  print('Treino', roc_train)

  y_proba_test = clf.predict_proba(X_test)
  roc_test = roc_auc_score(y_test, y_proba_test[:, 1])
  print('Test', roc_test)
  print('='*80)
```

```python
#a acuracia do RandomForestClassifier com a preditiva

clf=RandomForestClassifier()
clf.fit(X_train, y_train)
y_pred=clf.predict(X_test)
```

```python
#matriz da confusão com o plot de map

cm=confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='g')
plt.xlabel('Predição( 1 ou 0)')
plt.ylabel('Real ( 1 ou 0)')
```
